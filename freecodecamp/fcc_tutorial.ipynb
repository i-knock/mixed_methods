## Histograms

```r
library(datasets)
head(iris)
```

```r
hist(iris$Sepal.Length)
```

```r
table(iris$Species)
```

So the the "multi-plots", you can use `par(mfrow  = c(x, y))`
So if you want 3 plots on top of each other, you can just use:
`par(mfrow = c(3, 1))`

```r
par(mfrow = c(3, 1))

# First hist
hist(iris$Petal.Width [iris$Species == "setosa"],
    xlim = c(0, 3),
    breaks = 9,
    main = "Petal Widh for Setosa",
    xlab = "",
    col = "#14bea8")

# Second hist
hist(iris$Petal.Width [iris$Species == "versicolor"],
    xlim = c(0, 3),
    breaks = 9,
    main = "Petal Widh for versicolor",
    xlab = "",
    col = "purple")

# Third hist
hist(iris$Petal.Width [iris$Species == "virginica"],
    xlim = c(0, 3),
    breaks = 9,
    main = "Petal Widh for virginica",
    xlab = "",
    col = "red")
```

# Statistics
## Selecting cases
2 options:
- by *category*
- by *value*

```r
summary(iris$Petal.Length)
```

```r
# By category
hist(iris$Petal.Length[iris$Species == "versicolor"],
    main = "Petal Length: Versicolor")
```

```r
par(c(2, 1))

# By value
hist(iris$Petal.Length[iris$Petal.Length < 2],
    main = "Petal Length < 2")

# Multiple conditions
hist(iris$Petal.Length[
    iris$Species == "versicolor" & iris$Petal.Length < 5.5],
    main = "Versicolor and Petal Length: < 5.5")
```

```r
# For creating a sub-selection
i.setosa <- iris[iris$Species == "setosa", ]
summary(i.setosa)
```

# Data formats
- Data types: I.e. int, string, bools, etc.


```r
n1 <- 15  # double by default
typeof(n1)

n2 <- 1.5
typeof(n2)

c1 <- "c"
typeof(c1)

c2 <- "a string of text"
typeof(c2)

l1 <- TRUE # i.e. bool
l1 <- T
typeof(l1)
```


- Data structures:
  - Vector -> `c(...)`
    - 1-dim array
    - Scalars are also vectors
    - All same data type


```r
v1 <- c(1, 2, 3)
typeof(v1)
is.vector(v1)
```

  - Matrix
    - 2-dim array
    - All same length
    - All same data class


```r
m1 <- matrix(c(T,T,F,T,F,F), nrow = 2) #automatic shaping
m1

m2 <- matrix(c(
    'a', 'b',
    'c', 'd'
    ),
    nrow = 2
)
m2

m3 <- matrix(c(
    'a', 'b',
    'c', 'd'
    ),
    nrow = 2,
    byrow = T
)
m3
```

  - Array:
    - Matrix with 3+ dimensions

```r
a1 <- array(c(1:24), c(4, 3, 2))
a1
```

  - Data frame
    - Vectors of multiple types
    - All same length (keep in mind that "none" is an option)
    - Basically, a spreadsheet

```r
vNumeric   <- c(1, 2, 3)
vCharacter <- c("a", "b", "c")
vLogical   <- c(T, F, T)

dfa <- cbind(vNumeric, vCharacter, vLogical) # useful for adding columns
dfa <- rbind(dfa, c(4, 'd', F)) #useful for adding rows
dfa  # Matrix of one data type

df <- as.data.frame(dfa)
df  # Makes a data frame with three different data types

df[c("vNumeric", "vLogical")]
attributes(df)
```

```r
vec_char <- c("apple", "banana", "mandarin", "melon")
vec_num <- c(1, 5, 6, 3)

dat_fruit <- data.frame(name = vec_char, count = vec_num)
dat_fruit
```

  - List
    - Very flexible -> you can put whatever you want in them

```r
o1 <- c(1, 2, 3)
o2 <- c("a", "b", "c", "d")
o3 <- c(T, F, T, T, F)

list1 <- list(o1, o2, o3)
list1
```
